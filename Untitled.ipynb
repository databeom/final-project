```python
import pandas as pd
df = pd.read_csv('online_retail_listing.csv',sep=';',encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.groupby(['Customer ID','Invoice']).StockCode.count().describe()
```

```python
from datetime import datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'],format='%d.%m.%Y %H:%M')
df['date'] = df['InvoiceDate'].dt.strftime("%Y-%m")
df['date']
```

```python
# df.set_index("Customer ID", inplace=True)
FirstOrder = df.groupby(df.index)["InvoiceDate"].min()
FirstOrder = FirstOrder.dt.strftime("%Y-%m")
df["FirstOrder"] = FirstOrder
df.reset_index(inplace=True)
```

```python
grouped = df.groupby(["FirstOrder", "date"])

cohorts = grouped["Customer ID"].nunique()
cohorts = cohorts.reset_index()
cohorts.head()
```

```python
cohorts.rename({"Customer ID": "TotalUsers"}, axis=1, inplace=True)
cohorts.head()
```

```python
each_period = cohorts["FirstOrder"].value_counts().sort_index()
# 기간별 레이블 붙이기 
cohortperiod = []

for x in each_period:
    for y in range(x):
        cohortperiod.append(y)
```
