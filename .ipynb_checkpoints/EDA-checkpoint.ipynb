- 연령대 별

```python
df_living_popul = df.melt(id_vars=['상권_구분_코드_명','상권_코드_명'],
        value_vars=['연령대_10_유동인구_수','연령대_20_유동인구_수','연령대_30_40_유동인구_수','연령대_50_60_유동인구_수'],
        var_name = '연령',
       value_name = '유동인구').drop_duplicates().reset_index(drop=True)
```

```python
df_living_popul.loc[df_living_popul['연령']=='연령대_10_유동인구_수','연령'] = '10대'
df_living_popul.loc[df_living_popul['연령']=='연령대_20_유동인구_수','연령'] = '20대'
df_living_popul.loc[df_living_popul['연령']=='연령대_30_40_유동인구_수','연령'] = '30_40대'
df_living_popul.loc[df_living_popul['연령']=='연령대_50_60_유동인구_수','연령'] = '50_60대'
```

```python
df_living_popul
```

```python
df_resident_popul = df.melt(id_vars=['상권_구분_코드_명','상권_코드_명'],
        value_vars=['연령대_10_상주인구_수','연령대_20_상주인구_수','연령대_30_40_상주인구_수','연령대_50_60_상주인구_수'],
        var_name = '연령',
       value_name = '상주인구').drop_duplicates().reset_index(drop=True)
```

```python
df_resident_popul.loc[df_resident_popul['연령']=='연령대_10_상주인구_수','연령'] = '10대'
df_resident_popul.loc[df_resident_popul['연령']=='연령대_20_상주인구_수','연령'] = '20대'
df_resident_popul.loc[df_resident_popul['연령']=='연령대_30_40_상주인구_수','연령'] = '30_40대'
df_resident_popul.loc[df_resident_popul['연령']=='연령대_50_60_상주인구_수','연령'] = '50_60대'
```

```python
df_resident_popul
```

```python
df_sales = df.melt(id_vars=['상권_구분_코드_명','상권_코드_명'],
        value_vars=['연령대_10_매출_금액','연령대_20_매출_금액','연령대_30_40_매출_금액','연령대_50_60_매출_금액'],
        var_name = '연령',
       value_name = '매출액').drop_duplicates().reset_index(drop=True)
```

```python
df_sales.loc[df_sales['연령']=='연령대_10_매출_금액','연령'] = '10대'
df_sales.loc[df_sales['연령']=='연령대_20_매출_금액','연령'] = '20대'
df_sales.loc[df_sales['연령']=='연령대_30_40_매출_금액','연령'] = '30_40대'
df_sales.loc[df_sales['연령']=='연령대_50_60_매출_금액','연령'] = '50_60대'
```

```python
df_sales = df_sales.groupby(['상권_구분_코드_명','상권_코드_명','연령'],as_index=False).sum()
```

```python
df_num_sales = df.melt(id_vars=['상권_구분_코드_명','상권_코드_명'],
        value_vars=['연령대_10_매출_건수','연령대_20_매출_건수','연령대_30_40_매출_건수','연령대_50_60_매출_건수'],
        var_name = '연령',
       value_name = '매출건수').drop_duplicates().reset_index(drop=True)
```

```python
df_num_sales.loc[df_num_sales['연령']=='연령대_10_매출_건수','연령'] = '10대'
df_num_sales.loc[df_num_sales['연령']=='연령대_20_매출_건수','연령'] = '20대'
df_num_sales.loc[df_num_sales['연령']=='연령대_30_40_매출_건수','연령'] = '30_40대'
df_num_sales.loc[df_num_sales['연령']=='연령대_50_60_매출_건수','연령'] = '50_60대'
```

```python
df_num_sales = df_num_sales.groupby(['상권_구분_코드_명','상권_코드_명','연령'],as_index=False).sum()
```

```python
df = df_living_popul.merge(df_resident_popul,how='inner',on=['상권_구분_코드_명','상권_코드_명','연령'])
df = df.merge(df_sales,how='inner',on=['상권_구분_코드_명','상권_코드_명','연령'])
df = df.merge(df_num_sales,how='inner',on=['상권_구분_코드_명','상권_코드_명','연령'])
```
