# 라이브러리 설치

```python
import pandas as pd
import os
```

```python
pd.options.display.float_format = '{:.3f}'.format
pd.set_option('display.max_columns', None)
```

# 데이터 불러오기

```python
PATH = 'data/'

file_list= os.listdir(PATH)
csv_list = list()

for file in file_list:
    if file.split(".")[-1] == 'csv':
        csv_list.append(file)
```

```python
csv_list
```

```python
data = ['resident_popul', 'living_popul', 'indicator', 'apartment', 'income', 'area', 'facilitie', 'working_popul', 'store', 'estimated_sales']

for name, file in zip(data, csv_list):
    globals()[name] = pd.read_csv(PATH + file, encoding = "cp949")
```

# EDA

## 전체 데이터

### 데이터 확인

```python
for i in data:
  print(i)
  print('------------------------------------------------')
  globals()[i].info()
  print('================================================')
```

### 결측값 확인

```python
for i in data:
  print(i)
  print('------------------------------------------------')
  print(globals()[i].isnull().sum())
  print('================================================')
```

### 0인 값 확인

```python
for i in data:
  col = list(globals()[i].columns)
  print(i)
  print('------------------------------------------------------')
  for j in col:
    print(f'{j} : {len(globals()[i][globals()[i][j] == 0])}')
  print('======================================================')
```
